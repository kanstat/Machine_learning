employee salary prediction problem using decision tree.
decision tree--->>It is a tree-structured classifier, where internal nodes represent the features of a dataset, branches represent the decision rules and each leaf node represents the outcome.(https://www.javatpoint.com/machine-learning-decision-tree-classification-algorithm)

```python
import pandas as pd 
```

```python
df = pd.read_csv("/content/salaries.csv")
df
```

```python
inputs = df.drop('salary_more_then_100k',axis='columns')
```

```python
target = df['salary_more_then_100k']
```

```python
inputs
```

```python
target
```

```python
#ml algorithm works only on numbers so we have to convert columns into numbers by label hot encoding
from sklearn.preprocessing import LabelEncoder

```

```python
#create object of labelencoderclass
ob1_company = LabelEncoder()
ob2_job = LabelEncoder()
ob3_degree = LabelEncoder()
```

```python
#in inputs df create one more column
inputs['company_m'] = ob1_company.fit_transform(inputs['company'])
inputs['ob2_job'] = ob1_company.fit_transform(inputs['job'])
inputs['ob3_degree'] = ob1_company.fit_transform(inputs['degree'])
```

```python
#label encoded labels into numbers
inputs
```

```python
#drop labels
inputs_1 = inputs.drop(['company','job','degree'], axis = 'columns')
inputs_1
```

```python
#make model 
from sklearn import tree
```

```python
model= tree.DecisionTreeClassifier()
```

```python
#train model
model.fit(inputs_1,target)
```

```python
model.score(inputs_1,target)
```

```python
model.predict([[2,	2,	1]])
```
