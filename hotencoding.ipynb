# ***One hot encoding***


```python
import pandas as pd
url = "/content/homeprices.csv"
df = pd.read_csv(url)
df
```

```python
#using pandas to get dummy variables
dummy_var = pd.get_dummies(df.town)
dummy_var
```

```python
#merge or join dummy and original data frame
merged = pd.concat([df,dummy_var],axis='columns')
merged
```

```python

f = merged.drop(['town','west windsor'],axis='columns')
f
```

```python

X =f.drop('price',axis='columns')
X
```

```python
y= f.price
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict(X)
```

```python
model.score(X,y)
```
